```python
import numpy as np
import gzip
import pickle
import sys
import copy
import os.path
from copy import deepcopy

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from collections import defaultdict
agent_names = ['EWC_mnist', 'SI', 'GEM_1100']
#agent_names = ['EWC_mnist', 'SI', 'GEM_4000'] # for perm
hh = [10, 50, 100, 400]
model_names = ['MLP{0}SL'.format(h) for h in hh]
datasets = ['split_mnistb', 'split_mnistr', 'split']
#datasets = ['perm'] # use GEM_4000
cl_tasks = ['cl1','cl2', 'cl3']
tag = lambda data, cl : 'l1_{0}_{1}_aistats'.format(data, cl)

nested_dict = lambda: defaultdict(nested_dict)
d = nested_dict()

for aa in agent_names:
    for mm in model_names:
        for dd in datasets:
            for cl in cl_tasks:
                
                with open('../../continual_learning_baselines/outputs/{0}_{1}_{2}.pkl'.format(aa,
                                                                                              mm, 
                                                                                              tag(dd, cl)), 'rb') as input_file:
                    tmp = pickle.load(input_file)
                    for reg_coef, v in tmp['_accs'].items():
                        print('reg_coef:', reg_coef, 'agent:', aa, 'model:', mm, 'data:', dd, 'cl:', cl,
                              'mean: {:.3f}'.format(tmp['_accs'][reg_coef].mean()),
                              'std: {:.3f}'.format(tmp['_accs'][reg_coef].std()))

                        d[aa][mm][dd][cl] = tmp['_accs'][reg_coef]
```

```python
cl = 'cl2'
dd = 'split_mnistb'
for aa in agent_names:
    acc = []
    stds = []
    for mm in model_names:
        acc.append(d[aa][mm][dd][cl].mean())
        stds.append(d[aa][mm][dd][cl].std())
            
    print('agent:', aa, 'data:', dd, 'cl:', cl,
          'mean: {:.1f}'.format(np.mean(acc)),
          'std: {:.1f}'.format(np.sqrt(np.mean([s**2 + m**2 for m, s in zip(acc, stds)]) - np.mean(acc)**2)),
          'median: ${:.1f}$'.format(np.median(acc)),
          '\scalebox{{0.75}}{{$({:.1f}, {:.1f})$}}'.format(np.max(acc), np.min(acc)))
```
